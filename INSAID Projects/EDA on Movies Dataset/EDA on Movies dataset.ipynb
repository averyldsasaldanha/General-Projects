````python
bool_series = pd.isnull(update_moviedata["Revenue (Millions)"])    # Method to find out the NaN values
update_moviedata[bool_series]
````

````python
bool_series = pd.isnull(update_moviedata["Metascore"])                  # Method to find out the NaN values
update_moviedata[bool_series]
````

````python
# Method to find the Mean and Median of the 2 columns which have null values and then print the same

Revenue_Mean = update_moviedata['Revenue (Millions)'].mean()                               # 83.05161882893223
print (Revenue_Mean)
Revenue_Median = update_moviedata['Revenue (Millions)'].median()                           # 48.02             
print (Revenue_Median)

Metascore_Mean = update_moviedata['Metascore'].mean()                                      # 58.99465240641711
print (Metascore_Mean)
Metascore_Median = update_moviedata['Metascore'].median()                                  # 60.0          
print (Metascore_Median)
````

As the **Mean value** is very close to the **Median Value** we will replace the 2 null value columns with the Median values as below:

````python
# Replacing the null values with the Median value

update_moviedata['Revenue (Millions)'].fillna(Revenue_Median, inplace = True)
update_moviedata['Metascore'].fillna(Metascore_Median, inplace = True)
````

# **8. Exploratory Data Analysis**
---

**Exploratory Data Analysis(EDA)** is an **approach to analyzing data sets** to **summarize** their **main characteristics**, often with **visual methods**. 


- It includes **cleaning, munging, combining, reshaping, slicing, dicing, and transforming data** for **analysis** purpose.


- The **primary goal of EDA** is to **maximize the analyst's insight into a data set** and into the **underlying structure of a data set**, while **providing all of the specific items** that an **analyst would want to extract from a data set**, such as:
 <br><br> 
  - A list of **outliers**.<br><br>
  
  - A **good-fitting model**.  <br><br>
  
  - **Estimates for parameters**.<br><br>
  
  - **Uncertainties** for those **estimates**.<br><br>
  
  - A **ranked list** of **important factors**.<br><br>
  
  - **Conclusions** as to whether **individual factors are statistically significant**.<br><br>
  
  - A **sense of robustness of conclusions**.<br><br>
  
  - **Optimal settings**.<br><br>

**2. ANALYSIS OF THE MOVIE WITH THE HIGHEST AND THE LOWEST EARNED REVENUE (MILLIONS)**
(Here we will assume that the Revenue is in the currency USD)

````python
def find_minmax(x):
    
    min_index = update_moviedata[x].idxmin()                  # use the function 'idmin' to find the index of lowest Revenue movie.                  
    high_index = update_moviedata[x].idxmax()                 # use the function 'idmax' to find the index of Highest Revenue movie.
    high = pd.DataFrame(update_moviedata.loc[high_index,:])
    low = pd.DataFrame(update_moviedata.loc[min_index,:])

#print the movie with high and low profit
    print("Movie Which Has Highest "+ x + " : ",update_moviedata['Title'][high_index])
    print("Movie Which Has Lowest "+ x + "  : ",update_moviedata['Title'][min_index])
    return pd.concat([high,low],axis = 1)

#call the find_minmax function.
find_minmax('Revenue (Millions)')
````

The column with index 50 shows the highest Revenue made by a movie and the column with index 27 shows the movie with the highest loss in Revenue in this dataset.

As we can see that **Star Wars: Episode VII - The Force Awakens** movie directed by **J.J. Abrams** earns the highest revenue in all, making 936.63 revenue in this dataset.  

The movie that has the least revenue in this dataset is **Dead Awake**, with a revenue of 0.01 directed by **Phillip Guzman**.

````python
#top 10 movies which made highest revenue.
#sort the 'Revenue (Millions)' column in decending order and store it in the new dataframe.

info = pd.DataFrame(update_moviedata['Revenue (Millions)'].sort_values(ascending = False))
info['Title'] = update_moviedata['Title']
data = list(map(str,(info['Title'])))

#extract the top 10 movies with high revenue data from the list and dataframe.
x = list(data[:10])
y = list(info['Revenue (Millions)'][:10])

#make the point plot and setup the title and labels.
ax = sns.pointplot(x=y,y=x)
sns.set(rc={'figure.figsize':(10,5)})
ax.set_title("Top 10 High Revenue Movies",fontsize = 15)
ax.set_xlabel("Revenue (Millions)",fontsize = 13)
ax.set_ylabel("Movies",fontsize = 13)
sns.set_style("darkgrid")
````

````python
def plot_correlation_map( df ):
    corr = df.corr()
    _ , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 240 , 10 , as_cmap = True )
    _ = sns.heatmap(corr,cmap = cmap,square=True, cbar_kws={ 'shrink' : .9 }, ax=ax, annot = True, annot_kws = { 'fontsize' : 12 })
````
